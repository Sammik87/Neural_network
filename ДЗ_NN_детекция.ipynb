```python
import numpy as np
import os
import shutil
from pathlib import Path
from glob import glob
```

<font size = 4><div style="text-align: justify">**YOLOv3:**</div></font>

```python
path_model_yolov3 = 'D:\\Стас\\Ученье - Свет\\Оценка\\Аспирантура\\Видео по NN, Data mining, ABC и т.д\\Распределение Гаусса, регрессия\\Python\\4. Нейронные сети\\Юлия\\YOLO'
path_model_yolov3
```

```python
from imageai.Detection import ObjectDetection
import os

detector = ObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath(os.path.join(path_model_yolov3, "yolov3.pt"))
detector.loadModel()
```

```python
execution_path = os.getcwd()
execution_path
```

```python
detections = detector.detectObjectsFromImage(input_image = os.path.join(execution_path, "image (1).jpg"),
                                             output_image_path = os.path.join(execution_path, "image_detected.png"),
                                             minimum_percentage_probability = 30)

for eachObject in detections:
    print(eachObject["name"] , " : ", eachObject["percentage_probability"], " : ", eachObject["box_points"] )
    print("--------------------------------")
```

```python
from PIL import Image
Image.open('image_detected.png')
```

```python
from imageai.Detection.Custom import DetectionModelTrainer

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory = "hololens")
trainer.setTrainConfig(object_names_array = ["hololens"],
                       batch_size = 8,
                       num_experiments = 30,
                       train_from_pretrained_model = "yolov3.pt")
trainer.trainModel()
```

```python
model_path = sorted(list(Path('hololens/models/').iterdir()))[-1]
model_path
```

```python
from imageai.Detection.Custom import CustomObjectDetection

detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath("hololens/models/yolov3_hololens_mAP-0.65372_epoch-27.pt")
detector.setJsonPath("hololens/json/hololens_yolov3_detection_config.json")
detector.loadModel()

detections = detector.detectObjectsFromImage(input_image = os.path.join(execution_path, "image (297).jpg"), 
                                             output_image_path = os.path.join(execution_path, "image_detected_2.jpg"))

for detection in detections:
    print(detection["name"], " : ", detection["percentage_probability"], " : ", detection["box_points"])
```
